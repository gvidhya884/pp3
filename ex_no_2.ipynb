```python
# Initial list of students (with duplicates)
students = ["vidhya", "kavi", "Divya", "dhanya", "dharani"]
print("Initial Student List:", students)
# Add a new student at the end
students.append("bavana")
print("After adding new student:", students)
# Create a backup copy of the list
backup_students = students.copy()
print("Backup List:", backup_students)
# Count how many times a student name appears
name = "divya"
print(name, "appears", students.count(name), "times")
# Merge transfer students list
transfer_students = ["vikasini", "sreema"]
students.extend(transfer_students)
print("After merging transfer students:", students)
# Find the position of a student
print("Position of Divya:", students.index("Divya"))
# Insert a priority student at a specific position
students.insert(1, "Priya")
print("After inserting priority student:", students)
# Remove the last enrolled student
removed_student = students.pop()
print("Removed Student:", removed_student)
# Remove a specific withdrawn student
students.remove("dharani")
print("After removing withdrawn student:", students)
# Display list in reverse order
students.reverse()
print("Reversed List:", students)
# Sort the list alphabetically
students.sort()
print("Sorted List:", students)
# Clear all records at the end of semester
students.clear()
print("Student List after clearing:", students)
```

```python
# Create a tuple to store product details
product = (101, "Laptop", "Electronics", 55000)
print("Product Details:", product)
# Access elements using indexing
print("Product ID:", product[0])
print("Product Name:", product[1])
print("Category:", product[2])
print("Price:", product[3])
# Tuple of product categories
categories = ("Electronics", "Clothing", "Electronics", "Grocery")
# Count how many times a category appears
print("Electronics count:", categories.count("Electronics"))
# Find index of a product name
products = ("Laptop", "Mobile", "Tablet", "Camera")
print("Index of Tablet:", products.index("Tablet"))
# Tuple slicing
print("Product name and category:", product[1:3])
# Convert tuple to list to update price
product_list = list(product)
product_list[3] = 60000  # updating price
product = tuple(product_list)
print("Updated Product Details:", product)
```
