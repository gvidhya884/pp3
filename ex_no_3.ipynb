```python
books = {
    101: {"title": "Python Basics", "author": "Guido", "available_copies": 5},
    102: {"title": "Data Science", "author": "James", "available_copies": 3},
    103: {"title": "AI Intro", "author": "Andrew", "available_copies": 7}
}
# Backup
backup_books = books.copy()
print("Backup:", backup_books)
# Initialize categories
categories = dict.fromkeys(["Science", "Arts", "Engineering"], 0)
print("Categories:", categories)
# Safe retrieval
print("Copies of book 101:", books.get(101)["available_copies"])
print("Copies of book 105:", books.get(105, "Book not found"))
# Display all books
print("All Books:")
for book_id, details in books.items():
    print(book_id, details)
# Extract book IDs and copy counts
print("Book IDs:", list(books.keys()))
print("Available Copies:", [b["available_copies"] for b in books.values()])
# Remove a book
books.pop(102)
print("After removing book 102:", books)
# Remove last added book
books.popitem()
print("After popitem:", books)
# Add missing category
categories.setdefault("Medical", 0)
print("Updated Categories:", categories)
# Merge new shipment
new_books = {
    104: {"title": "ML", "author": "Tom", "available_copies": 4}
}
books.update(new_books)
print("After update:", books)
# Clear all records
books.clear()
print("Books after clearing:", books)
```

```python
registered_students = {1, 2, 3, 4, 5}
workshop_A = {2, 3, 6}
workshop_B = {3, 4, 7}
volunteers = {1, 3}
# Add new registration
registered_students.add(6)
# Backup
backup = registered_students.copy()
# Registered but not in workshop A
print("Not in Workshop A:", registered_students.difference(workshop_A))
# Remove cancelled student
registered_students.remove(5)
# Safe remove
registered_students.discard(10)
# Common students
print("Common in A & B:", workshop_A.intersection(workshop_B))
# Volunteers part of registered?
print("Volunteers registered:", volunteers.issubset(registered_students))
# Keep only common attendees
workshop_A.intersection_update(workshop_B)
print("Workshop A after update:", workshop_A)
# Immutable group
final_volunteers = frozenset(volunteers)
# Attended only one workshop
print("Only one workshop:", workshop_A.symmetric_difference(workshop_B))
# Merge workshops
print("All participants:", workshop_A.union(workshop_B))
# Remove random volunteer
print("Removed volunteer:", volunteers.pop())
# Clear all
registered_students.clear()
print("Registered after clear:", registered_students)
```

```python
def count_vowels(s):
    vowels = "aeiou"
    count = 0
    for ch in s.lower():
        if ch in vowels:
            count += 1
    return count
print(count_vowels("Education"))
```

```python
students = {
    "Anu": [78, 85, 90],
    "vidhya": [60, 72, 68],
    "dhanya": [88, 91, 84],
    "Divya": [45, 50, 40]
}
def calculate_average(marks):
    return sum(marks) / len(marks)
results = {}
for name, marks in students.items():
    avg = round(calculate_average(marks), 2)
    status = (lambda x: "Pass" if x >= 50 else "Fail")(avg)
    results[name] = (avg, status)
print("Passed Students:")
for name, data in results.items():
    if data[1] == "Pass":
        print(name, data)
topper = max(results, key=lambda x: results[x][0])
print("Topper:", topper)
```

```python
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)
def sum_digits(n):
    if n == 0:
        return 0
    return n % 10 + sum_digits(n // 10)
def max_digit(n):
    if n < 10:
        return n
    return max(n % 10, max_digit(n // 10))
def reverse_number(n, rev=0):
    if n == 0:
        return rev
    return reverse_number(n // 10, rev * 10 + n % 10)
num = 1234
print("Digits:", count_digits(num))
print("Sum:", sum_digits(num))
print("Max Digit:", max_digit(num))
print("Reverse:", reverse_number(num))
```
